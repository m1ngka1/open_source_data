```python
import datetime
import pandas as pd
import numpy as np
import qlib
from qlib.data import D

from qlib.config import C
```

```python
qlib.init(provider_uri='~/.qlib/qlib_data/cn_data')
```

```python
D.calendar(start_time='2010-01-01', end_time='2017-12-31', freq='day')[:2]
```

```python
D.instruments(market='all')
```

```python
instruments = D.instruments(market='csi300')
ins_list = D.list_instruments(instruments=instruments, start_time='2010-01-01', end_time='2017-12-31', as_list=True)
print(len(ins_list), ins_list[:5])
```

```python
# Set a different backend
C.joblib_backend = 'threading'  # or 'sequential'

fields = ['$close', '$volume', 'Ref($close, 1)', 'Mean($close, 3)', '$high-$low']

df = D.features(ins_list[:5], fields, start_time='2025-01-01', end_time='2025-06-18', freq='day')
```
